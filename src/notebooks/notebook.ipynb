```python
database_client = SQLModelClient(database_url="sqlite:////Users/kunmi/Documents/tagging_system/data/database/dev_1.db")

with database_client as db:
  res = db.execute(
    """
    select [as].*, a.*, t.name as tag_name from asset a
      INNER JOIN asset_snapshot as [as]
        on a.id = [as].asset_id
      INNER JOIN asset_tag as at
        ON a.id = at.asset_id
      INNER JOIN tag as t
        ON at.tag_id = t.id
      LIMIT 500;
"""
  )

  res = res.fetchall()
```
